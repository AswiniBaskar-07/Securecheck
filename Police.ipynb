```python
#Create Table
create_table_query = """ 
CREATE TABLE IF NOT EXISTS policelog (
    stop_date DATETIME,
    stop_time DATETIME,
    country_name VARCHAR(200),
    driver_gender VARCHAR(10),
    driver_age_raw INT,
    driver_age INT,
    driver_race VARCHAR(200),
    violation_raw VARCHAR(200),
    violation VARCHAR(200),
    search_conducted VARCHAR(200),
    search_type VARCHAR(200),
    stop_outcome VARCHAR(200),
    is_arrested BOOLEAN,
    stop_duration VARCHAR(200),
    drugs_related_stop BOOLEAN,
    vehicle_number VARCHAR(200)
)
"""
```

```python
mycursor.execute(create_table_query)
print("Table created sucessfully")
```

```python
#SQL insert Query
insert_query = """
INSERT INTO policelog (
    stop_date, stop_time, country_name, driver_gender, driver_age_raw, driver_age,
    driver_race, violation_raw, violation, search_conducted, search_type,
    stop_outcome, is_arrested, stop_duration, drugs_related_stop, vehicle_number
)
VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""
```

```python
#Process DataFrame 
df['stop_time'] = pd.to_datetime(df['stop_time'], errors='coerce').dt.time
df['stop_date'] = pd.to_datetime(df['stop_date'], errors='coerce')
```

```python
# Insert each row 
for index, row in df.iterrows():
    stop_datetime = None
    if pd.notnull(row['stop_date']) and pd.notnull(row['stop_time']):
        stop_datetime = f"{row['stop_date'].strftime('%Y-%m-%d')} {row['stop_time'].strftime('%H:%M:%S')}"

    mycursor.execute(insert_query, (
        row['stop_date'].strftime('%Y-%m-%d %H:%M:%S') if pd.notnull(row['stop_date']) else None,
        stop_datetime,
        row['country_name'],
        row['driver_gender'],
        row['driver_age_raw'],
        row['driver_age'],
        row['driver_race'],
        row['violation_raw'],
        row['violation'],
        row['search_conducted'],
        row['search_type'],
        row['stop_outcome'],
        row['is_arrested'],
        row['stop_duration'],
        row['drugs_related_stop'],
        row['vehicle_number']
    ))
```

```python
# Commit Changes
mydb.commit()
print("All records inserted successfully!")
```

```python
# Close Connection
mycursor.close()
mydb.close()
```
